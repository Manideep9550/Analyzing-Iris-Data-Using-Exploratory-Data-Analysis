```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
# Load the Iris dataset
iris = sns.load_dataset('iris')
```

```python
# Display the first few rows of the dataset to understand its structure
print(iris.head())
```

```python
# Summary statistics of the dataset
print(iris.describe())
```

```python
# Information about the dataset
print(iris.info())
```

```python
# Checking for missing values
print(iris.isnull().sum())
```

```python
# Exploring the distributions of each feature using histograms
plt.figure(figsize=(10, 6))
for i, feature in enumerate(iris.columns[:-1]):
    plt.subplot(2, 2, i + 1)
    sns.histplot(iris[feature], kde=True)
    plt.title(f'Distribution of {feature}')
plt.tight_layout()
plt.show()
```

```python
# Exploring pairwise relationships with scatter plots
sns.pairplot(iris, hue='species', diag_kind='kde')
plt.show()
```

```python
# Exploring correlations between features with a heatmap

# Convert the 'species' column to numeric values
iris['species'] = iris['species'].astype('category').cat.codes

# Calculate the correlation matrix
correlation_matrix = iris.corr()

# Display the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
